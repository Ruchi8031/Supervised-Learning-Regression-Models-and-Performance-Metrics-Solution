Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

  - Simple Linear Regression (SLR) is a statistical and machine learning technique used to model and analyze the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line.

  - It establishes a linear relationship of the form:  Y=β0​+β1​X+ε

     - Purpose:

  - To predict continuous outcomes based on input variables.

  - To understand the strength and direction of relationships between variables.

  - To analyze trends and patterns in data.

  - To support decision-making and forecasting.
    - Applications:

  - Sales forecasting

  - Salary prediction

  - House price estimation

  - Demand forecasting

Question 2: What are the key assumptions of Simple Linear Regression?

  - The validity of Simple Linear Regression depends on the following assumptions:

    - Linearity:

  - There exists a linear relationship between X and Y.

    - Independence:

  - Observations are independent of each other.

    - Homoscedasticity:

  - The variance of residuals remains constant across all levels of X.

    -   Normality of Residuals:

  - Error terms are normally distributed.

    - No Multicollinearity:

  - Since SLR uses only one predictor, multicollinearity does not exist.

    - Importance:

  - Violation of these assumptions leads to biased results and inaccurate predictions.

Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

  - Y= β0 + β1 X + ε
	​
  - Y = Dependent variable (output)

  - X = Independent variable (input)

  - β₀ = Intercept (value of Y when X = 0)

  - β₁ = Slope (change in Y per unit change in X)

  - ε = Error term (residual)

Question 5: What is the method of least squares in linear regression?


   - Definition:
  - The method of least squares is a mathematical optimization technique used to find the best-fitting regression line by minimizing the sum of squared residuals.

  - Objective Function: min∑(Yactual​−Ypredicted​)2
    - Purpose:

  - To minimize prediction error.

  - To obtain the most accurate estimates of slope and intercept.

  - Practical Importance:

  - Produces a stable and optimal regression model.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

  - Logistic Regression is a supervised machine learning algorithm used for binary classification problems.

    - Key Characteristics:

  - Uses sigmoid (logistic) function

  - Outputs probabilities between 0 and 1

    - Differences:
  - Linear Regression	Logistic Regression
  - Predicts continuous values	Predicts categorical outcomes
  - Uses straight-line equation	Uses sigmoid function
  - Output ∈ (−∞, +∞)	Output ∈ (0,1)
  - Used for regression	Used for classification

    - Example:

  - Email spam detection

  - Disease prediction

Question 7: Name and briefly describe three common evaluation metrics for regression models.

  - Mean Absolute Error (MAE):

  - Average of absolute differences between actual and predicted values.

  - Measures average prediction error.

  - Mean Squared Error (MSE):

  - Mean of squared prediction errors.

  - Penalizes large errors more strongly.

  - Root Mean Squared Error (RMSE):

  - Square root of MSE.

  - Provides error in original data units.

Question 8: What is the purpose of the R-squared metric in regression analysis?

  - R-squared (R²) measures the proportion of variance in the dependent variable explained by the model.

    - Range:

  - 0 ≤ R² ≤ 1

    - Interpretation:

  - R² = 0 → Model explains nothing

  - R² = 1 → Perfect model

    - Importance:

  - Evaluates model performance and goodness of fit.

  - Helps in model comparison.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Input data
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([2, 4, 6, 8, 10])

# Create and train model
model = LinearRegression()
model.fit(X, Y)

# Output
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

  - Intercept (β₀):

  - Value of Y when X = 0.

  - Represents baseline prediction.

  - Slope (β₁):

  - Indicates the rate of change of Y with respect to X.

  - Shows the direction and strength of the relationship.

  - Example: If Y = 5 + 3X:

  - Intercept = 5

  - Slope = 3 → Every 1-unit increase in X increases Y by 3 units.
